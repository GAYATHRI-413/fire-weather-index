```python
# ===============================
# Algerian Forest Fires - Classification Pipeline
# Models: RandomForest, LogisticRegression
# ===============================

# 1Ô∏è‚É£ Import libraries
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt
import json
import os

# 2Ô∏è‚É£ Load dataset
file_path = "Algerian_forest_fires_dataset.csv"
df = pd.read_csv(file_path)

# 3Ô∏è‚É£ Clean column names (remove spaces)
df.columns = [c.strip() for c in df.columns]

# 4Ô∏è‚É£ Convert numeric-like columns to numeric
num_cols = ['day','month','year','Temperature','RH','Ws','Rain',
            'FFMC','DMC','DC','ISI','BUI','FWI']
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')

# 5Ô∏è‚É£ Clean and normalize the target column
df['Classes'] = df['Classes'].astype(str).str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)

# Remove rows that look like headers or invalid
bad_vals = ['classes', 'nan', 'none', '']
df = df[~df['Classes'].isin(bad_vals)].copy()
df = df[df['Classes'].notna()]

# Keep only fire / not fire
df = df[df['Classes'].isin(['fire', 'not fire'])].copy()

# Map to binary target
df['target'] = (df['Classes'] == 'fire').astype(int)

print(f"‚úÖ Cleaned dataset shape: {df.shape}")

# 6Ô∏è‚É£ Define features and target
features = ['day','month','year','Temperature','RH','Ws','Rain',
            'FFMC','DMC','DC','ISI','BUI','FWI']
X = df[features]
y = df['target']

# 7Ô∏è‚É£ Handle missing values
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=features)

# 8Ô∏è‚É£ Train-test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y, test_size=0.2, random_state=42, stratify=y
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# 9Ô∏è‚É£ Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# üîü Train models
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=1000, random_state=42)

rf.fit(X_train_scaled, y_train)
lr.fit(X_train_scaled, y_train)

# 1Ô∏è‚É£1Ô∏è‚É£ Evaluate models
models = {"RandomForest": rf, "LogisticRegression": lr}
results = {}

for name, m in models.items():
    preds = m.predict(X_test_scaled)
    probs = m.predict_proba(X_test_scaled)[:, 1]
    acc = accuracy_score(y_test, preds)
    auc = roc_auc_score(y_test, probs)
    cr = classification_report(y_test, preds, output_dict=True)
    cm = confusion_matrix(y_test, preds)
    results[name] = {
        "accuracy": acc,
        "roc_auc": auc,
        "classification_report": cr,
        "confusion_matrix": cm.tolist()
    }

# 1Ô∏è‚É£2Ô∏è‚É£ Pick best model
best_name = max(results.items(), key=lambda kv: kv[1]['accuracy'])[0]
best_model = models[best_name]
print(f"üèÜ Best model: {best_name}")

# 1Ô∏è‚É£3Ô∏è‚É£ Save model pipeline
joblib.dump(
    {
        'model': best_model,
        'scaler': scaler,
        'imputer': imputer,
        'features': features
    },
    "best_model_pipeline.joblib"
)

# 1Ô∏è‚É£4Ô∏è‚É£ Save feature importances (for RandomForest)
fi = pd.DataFrame({
    'feature': features,
    'importance': best_model.feature_importances_
}).sort_values('importance', ascending=False)
fi.to_csv("feature_importances.csv", index=False)

# 1Ô∏è‚É£5Ô∏è‚É£ Compute ROC data and save
fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(X_test_scaled)[:, 1])
np.save("roc_fpr.npy", fpr)
np.save("roc_tpr.npy", tpr)

# 1Ô∏è‚É£6Ô∏è‚É£ Save model results JSON
with open("model_results.json", "w") as f:
    json.dump(results, f)

# 1Ô∏è‚É£7Ô∏è‚É£ Save dataset summary JSON
summary = {
    "shape": df.shape,
    "columns": df.columns.tolist(),
    "classes": df['Classes'].value_counts().to_dict()
}
with open("dataset_summary.json", "w") as f:
    json.dump(summary, f)

# 1Ô∏è‚É£8Ô∏è‚É£ Save cleaned dataset
df.to_csv("Algerian_forest_fires_dataset_processed.csv", index=False)

# 1Ô∏è‚É£9Ô∏è‚É£ Show results summary
results_df = pd.DataFrame({
    k: {"accuracy": v["accuracy"], "roc_auc": v["roc_auc"]}
    for k, v in results.items()
}).T
print("\n=== Model Results Summary ===")
print(results_df)

# 2Ô∏è‚É£0Ô∏è‚É£ Confusion matrix and ROC curve plots
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, best_model.predict(X_test_scaled))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['not fire','fire'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Best Model")
plt.show()

# ROC Curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, best_model.predict_proba(X_test_scaled)[:, 1]):.3f}")
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.title("ROC Curve - Best Model")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

print("\n‚úÖ All artifacts saved successfully:")
print("- dataset_summary.json")
print("- Algerian_forest_fires_dataset_processed.csv")
print("- best_model_pipeline.joblib")
print("- model_results.json")
print("- feature_importances.csv")
print("- roc_fpr.npy, roc_tpr.npy")
```
